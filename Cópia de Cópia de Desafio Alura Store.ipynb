```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()

```

```python
#1. Análise do faturamento

faturamentoloja1 = sum(loja.Preço)
faturamentoloja2 = sum(loja2.Preço)
faturamentoloja3 = sum(loja3.Preço)
faturamentoloja4 = sum(loja4.Preço)

print(f"Faturamento: Loja 1 - R$ {faturamentoloja1:,.2f} \nFaturamento: Loja 2 - R$ {faturamentoloja2:,.2f} \nFaturamento: Loja 3 - R$ {faturamentoloja3:,.2f} \nFaturamento: Loja 4 - R$ {faturamentoloja4:,.2f}")
print(f'Faturamento Total: =  R$ {faturamentoloja1 + faturamentoloja2 + faturamentoloja3 + faturamentoloja4:,.2f}')
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
lojas = ['loja1', 'loja2', 'loja3', 'loja4']
faturamento = [faturamentoloja1, faturamentoloja2, faturamentoloja3, faturamentoloja4]
# Criando o gráfico de barras
plt.bar(lojas, faturamento)

# Adicionando rótulos e título
plt.xlabel('lojas')
plt.ylabel('faturamento')
plt.title('Gráfico de Barras Simples')

# Exibindo o gráfico
plt.show()

```

```python
#2. Vendas por Categoria

vendas_cat = loja.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print(f'Vendas por categoria Loja 1: \n{vendas_cat} \n')

vendas_cat2 = loja2.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print(f'Vendas por categoria Loja 2: \n{vendas_cat2} \n')

vendas_cat3 = loja3.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print(f'Vendas por categoria Loja 3: \n{vendas_cat3} \n')

vendas_cat4 = loja4.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print(f'Vendas por categoria Loja 4: \n{vendas_cat4} \n')

junta_cat = vendas_cat +  vendas_cat2 + vendas_cat3 + vendas_cat4

soma_cat = junta_cat.groupby('Categoria do Produto').sum().sort_values(ascending=False)

print(f'Vendas total das Categoria nas 4 Loja: \n{soma_cat} \n')

# Gráfico de barras para o total de vendas por categoria nas 4 lojas
soma_cat.plot(kind='bar', color='blue')
plt.title('Total de Vendas por Categoria (4 Lojas)')
plt.xlabel('Categoria do Produto')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
#3. Média de Avaliação das Lojas

avaliacao1 = loja["Avaliação da compra"].mean().item()
avaliacao2 = loja2["Avaliação da compra"].mean().item()
avaliacao3 = loja3["Avaliação da compra"].mean().item()
avaliacao4 = loja4["Avaliação da compra"].mean().item()

print(f'Média de avaliação das lojas: \nLoja 1 = {avaliacao1:,.2f} \nLoja 2 = {avaliacao2:,.2f} \nLoja 3 = {avaliacao3:,.2f} \nLoja 4 = {avaliacao4:,.2f} \n')


```

```python
#4. Produtos Mais e Menos Vendidos

produto_mais1 = loja["Produto"].value_counts()
produto_mais2 = loja2["Produto"].value_counts()
produto_mais3 = loja3["Produto"].value_counts()
produto_mais4 = loja4["Produto"].value_counts()
print(f'Produtos mais vendidos \nLoja 1 {produto_mais1} \n\nLoja 2 {produto_mais2} \n\nLoja 3 {produto_mais3} \n\nLoja 4 {produto_mais4}\n\n')

soma_produto = (produto_mais1 + produto_mais2 + produto_mais3 + produto_mais4).sort_values(ascending=False)
print(f'Total de produtos das 4 lojas: \n{soma_produto}')


```

```python
#5. Frete Médio por Loja

frete_medio = loja['Frete'].mean().item()
frete_medio2 = loja2['Frete'].mean().item()
frete_medio3 = loja3['Frete'].mean().item()
frete_medio4 = loja4['Frete'].mean().item()

print(f'Média de frete das lojas \nLoja 1 = {frete_medio:,.2f} \nLoja 2 = {frete_medio2:,.2f} \nLoja 3 = {frete_medio3:,.2f} \nLoja 4 = {frete_medio4:,.2f}')

# Dados para o gráfico
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias_frete = [frete_medio, frete_medio2, frete_medio3, frete_medio4]

# Criando o gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(nomes_lojas, medias_frete, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# Adicionando rótulos e título
plt.xlabel('Lojas')
plt.ylabel('Média de Frete')
plt.title('Média de Frete por Loja')

# Adicionando os valores acima das barras
for i, v in enumerate(medias_frete):
    plt.text(i, v + 0.1, f'R$ {v:.2f}', ha='center', va='bottom')

# Exibindo o gráfico
plt.tight_layout()
plt.show()

```
